```python

import pandas as pd

# Specify the file path
file_path = 'sleep_day_data_cleaned.csv'  # Replace 'your_file_path.csv' with the actual file path

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame to verify the import
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot Histogram for TotalMinutesAsleep
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalMinutesAsleep'], bins=30, kde=True, color='skyblue')
plt.title('Histogram for TotalMinutesAsleep')
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot Bar Plot for TotalSleepRecords
plt.figure(figsize=(8, 6))
sns.barplot(x='Id', y='TotalSleepRecords', data=df, palette='pastel')
plt.title('Bar Plot for TotalSleepRecords')
plt.xlabel('ID')
plt.ylabel('Total Sleep Records')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot Box Plot for TotalTimeInBed
plt.figure(figsize=(10, 6))
sns.boxplot(x='TotalTimeInBed', data=df, color='lightgreen')
plt.title('Box Plot for TotalTimeInBed')
plt.xlabel('Total Time In Bed')
plt.show()
```

```python
import matplotlib.pyplot as plt


# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plot TotalMinutesAsleep - Histogram
axes[0, 0].hist(df['TotalMinutesAsleep'], bins=30, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Histogram for TotalMinutesAsleep')
axes[0, 0].set_xlabel('Total Minutes Asleep')
axes[0, 0].set_ylabel('Frequency')

# Plot TotalSleepRecords - Bar Plot
axes[0, 1].bar(df['Id'], df['TotalSleepRecords'], color='lightcoral')
axes[0, 1].set_title('Bar Plot for TotalSleepRecords')
axes[0, 1].set_xlabel('ID')
axes[0, 1].set_ylabel('Total Sleep Records')

# Plot TotalTimeInBed - Box Plot
axes[1, 0].boxplot(df['TotalTimeInBed'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
axes[1, 0].set_title('Box Plot for TotalTimeInBed')
axes[1, 0].set_xlabel('Total Time In Bed')

# Hide the empty subplot
axes[1, 1].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```
