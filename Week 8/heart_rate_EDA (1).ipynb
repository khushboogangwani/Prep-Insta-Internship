```python

import pandas as pd

# Specify the file path
file_path = 'heart_rate_cleaned.csv'  # Replace 'your_file_path.csv' with the actual file path

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame to verify the import
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Plot Histogram of Heartrate
plt.figure(figsize=(10, 6))
sns.histplot(df['Heartrate'], bins=30, kde=True, color='skyblue')
plt.title('Histogram of Heartrate')
plt.xlabel('Heartrate')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Day' to datetime for proper plotting
df['Day'] = pd.to_datetime(df['Day'])

# Plot Line Plot of Daily Mean Heartrate Over Time
plt.figure(figsize=(12, 8))
sns.lineplot(x='Day', y='DailyMeanHeartrate', data=df, marker='o', color='green')
plt.title('Daily Mean Heartrate Over Time')
plt.xlabel('Date')
plt.ylabel('Daily Mean Heartrate')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Plot Scatter Plot of Time vs. Heartrate with Color by Daily Mean Heartrate
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Time', y='Heartrate', data=df, hue='DailyMeanHeartrate', palette='viridis', alpha=0.7)
plt.title('Scatter Plot: Time vs. Heartrate')
plt.xlabel('Time')
plt.ylabel('Heartrate')
plt.legend(title='Daily Mean Heartrate', loc='upper right')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the relevant columns for the pair plot
pair_plot_columns = ['Heartrate', 'DailyMeanHeartrate']

# Create Pair Plot with Scattered Values
sns.pairplot(df[pair_plot_columns], kind='scatter', markers='.', diag_kind='kde', palette='husl')
plt.suptitle('Pair Plot with Scattered Values', y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns for box plots
numeric_columns = df.select_dtypes(include='number').columns

# Create Box Plots for All Numeric Columns
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[column], color='skyblue')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```
