# **Exploratory Data Analysis Report:**

1st daily data set

```python
# import packages and alias
import numpy as np # data arrays
import pandas as pd # data structure and data analysis
import matplotlib as plt # data visualization
import datetime as dt # date time
import seaborn as sns
```

```python
df= pd.read_csv("/content/daily_activity_cleaned.csv")
df1= pd.read_csv("/content/daily_calories_cleaned.csv")
df2= pd.read_csv("/content/daily_intensities_cleaned.csv")
```

```python
df
```

```python
df1
```

```python
df2
```

```python
merged_df = pd.merge(df, df1, on="Id", how="inner")
```

```python
merged_df = pd.merge(merged_df, df2, on="Id", how="inner")
```

```python
merged_df
```

```python


# Check for missing values
missing_values = merged_df.isnull().sum()
merged_df = merged_df.dropna()  # Drop rows with missing values

# Display basic statistics
merged_df.describe()
```

```python
merged_df.columns
```

Physical Activity:
Explore the distribution of daily step counts and active minutes.

```python
import matplotlib.pyplot as plt

# Visualize step count distribution
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['TotalSteps'], bins=30, kde=True)
plt.title('Distribution of Daily Step Counts')
plt.xlabel('Total Steps')
plt.ylabel('Frequency')
plt.show()

```

```python
# 2. Heart Rate

plt.figure(figsize=(10, 6))
sns.histplot(merged_df['SedentaryMinutes_x'], bins=30, kde=True, color='salmon', label='Resting Heart Rate')
plt.title('Distribution of Resting Heart Rate')
plt.xlabel('Resting Heart Rate')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Bivariate EDA with scatter plots
sns.scatterplot(x='TotalSteps', y='Calories_x', data=merged_df)
plt.title('Scatter Plot: Total Steps vs Calories_x')
plt.show()

sns.scatterplot(x='VeryActiveMinutes_x', y='Calories_x', data=merged_df)
plt.title('Scatter Plot: Very Active Minutes vs Calories_x')
plt.show()

```

```python
# Bivariate EDA with pair plots
columns_of_interest = ['TotalSteps', 'VeryActiveMinutes_x', 'LightlyActiveMinutes_x', 'SedentaryMinutes_x', 'Calories_x']
sns.pairplot(merged_df[columns_of_interest])
plt.suptitle('Pair Plot of Selected Columns', y=1.02)
plt.show()
```

```python
#This will help you visualize the correlation between numerical columns.
correlation_matrix = merged_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python

plt.figure(figsize=(12, 8))

# Box Plot: Very Active Minutes vs Calories_x
sns.boxplot(x='VeryActiveMinutes_x', y='Calories_x', data=merged_df, width=0.5, palette='Set3')
plt.title('Box Plot: Very Active Minutes vs Calories_x')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.regplot(x='TotalSteps', y='Calories_x', data=merged_df)
plt.title('Regression Plot: Total Steps vs Calories_x')
plt.show()
```

```python
# Joint Plot: Total Steps vs Calories_x
sns.jointplot(x='TotalSteps', y='Calories_x', data=merged_df, kind='scatter')
plt.suptitle('Joint Plot: Total Steps vs Calories_x', y=1.02)
plt.show()
```

```python
# Hexbin Plot: Total Steps vs Calories_x
sns.jointplot(x='TotalSteps', y='Calories_x', data=merged_df, kind='hex', gridsize=20)
plt.suptitle('Hexbin Plot: Total Steps vs Calories_x', y=1.02)
plt.show()
```

```python
# Pair Grid: Selected columns
columns_of_interest = ['TotalSteps', 'VeryActiveMinutes_x', 'LightlyActiveMinutes_x', 'Calories_x']
sns.pairplot(merged_df[columns_of_interest])
plt.suptitle('Pair Grid of Selected Columns', y=1.02)
plt.show()
```
