```python
# import packages and alias
import numpy as np # data arrays
import pandas as pd # data structure and data analysis
import matplotlib as plt # data visualization
import datetime as dt # date time
import seaborn as sns
```

```python
df= pd.read_csv("/content/hourly_calories_cleaned.csv")
df1= pd.read_csv("/content/hourly_intensities_cleaned.csv")
```

```python
df
```

```python
df1
```

```python
merged_df = pd.merge(df, df1, on="Id", how="inner")
```

```python
merged_df
```

```python
merged_df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(x='ActivityDay_x', data=merged_df)
plt.title('Count of ActivityDay_x')
plt.xticks(rotation=45)
plt.show()
```

```python
#Univariate analysis for 'ActivityHour_x':

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(merged_df['ActivityHour_x'], bins=24, kde=True)
plt.title('Histogram of ActivityHour_x')
plt.xlabel('ActivityHour_x')
plt.ylabel('Frequency')
plt.show()
```

```python
#Univariate analysis for 'ActivityDay_y':
plt.figure(figsize=(10, 6))
sns.countplot(x='ActivityDay_y', data=merged_df)
plt.title('Count of ActivityDay_y')
plt.xticks(rotation=45)
plt.show()
```

```python
#Univariate analysis for 'ActivityHour_y':
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['ActivityHour_y'], bins=24, kde=True)
plt.title('Histogram of ActivityHour_y')
plt.xlabel('ActivityHour_y')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.boxplot(x=merged_df['Calories'])
plt.title('Box Plot: Calories')
plt.xlabel('Calories')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.violinplot(x=merged_df['TotalIntensity'], inner='quartile', palette='Set2')
plt.title('Violin Plot: TotalIntensity')
plt.xlabel('TotalIntensity')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.kdeplot(merged_df['AverageIntensity'], fill=True, palette='husl')
plt.title('Kernel Density Plot: AverageIntensity')
plt.xlabel('AverageIntensity')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.scatterplot(x='ActivityHour_x', y='Calories', data=merged_df, hue='ActivityDay_x', palette='viridis')
plt.title('Scatter Plot: ActivityHour_x vs Calories')
plt.xlabel('ActivityHour_x')
plt.ylabel('Calories')
plt.legend(title='ActivityDay_x')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.regplot(x='TotalIntensity', y='AverageIntensity', data=merged_df, scatter_kws={'s': 20}, line_kws={'color': 'red'})
plt.title('Regression Plot: TotalIntensity vs AverageIntensity')
plt.xlabel('TotalIntensity')
plt.ylabel('AverageIntensity')
plt.show()
```
