```python
#Filling The Null Values by Using Mean

years_to_fill = [str(year) for year in range(1960, 2017)]

for year in years_to_fill:
    mean_year = df1[year].mean()
    df1[year].fillna(mean_year, inplace=True)
```

# **Merging The Datasets**

```python
life_exp_melted = df1.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Life Expectancy')
```

```python
life_exp_melted
```

```python
fertility_melted = df2.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Fertility Rate')
```

```python
fertility_melted
```

```python
country_population_melted = df3.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Country Population')
```

```python
country_population_melted
```

```python
merged_df = pd.merge(country_population_melted, life_exp_melted, on=['Country Name', 'Country Code', 'Year'])
```

```python
merged_df
```

```python
merged_df = pd.merge(merged_df, fertility_melted, on=['Country Name', 'Country Code', 'Year'])
```

```python
merged_df
```

```python
Dataframe = pd.merge(merged_df, df4, on='Country Code')
```

```python
Dataframe
```

```python
final_columns = [
    'Country Name', 'Country Code', 'Year',
     'Country Population','Life Expectancy', 'Fertility Rate',
    'Region', 'IncomeGroup'
]

Dataframe = Dataframe[final_columns]
```

```python
Dataframe
```

```python
import plotly.express as px

fig = px.scatter(
    Dataframe,
    x='Fertility Rate',
    y='Life Expectancy',
    size='Country Population',
    color='Region',
    animation_frame='Year',
    animation_group='Country Name',
    hover_name='Country Name',
    log_x=True,
    range_y=[25, 95],
    height=600,
    size_max=100,
    title='Fertility Rate Vs Life Expectancy',

)

fig.update_layout(
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
    title_x=0.5,
    title_y=0.9,


)

fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 600

fig.show()
```

# **The Final Output**

```python
import plotly.express as px


region_order = Dataframe['Region'].unique()

fig = px.scatter(
    Dataframe,
    x='Fertility Rate',
    y='Life Expectancy',
    size='Country Population',
    color='Region',
    animation_frame='Year',
    animation_group='Country Name',
    hover_name='Country Name',
    log_x=True,
    range_y=[25, 95],
    height=600,
    size_max=100,
    title='Fertility Rate vs Life Expectancy',
    labels={'Fertility Rate': 'Fertility Rate',
            'Life Expectancy': 'Life Expectancy'},
    symbol='Region',
    opacity=0.8,
    color_discrete_sequence=px.colors.qualitative.Set1,
    category_orders={'Region': region_order},
)


fig.update_layout(
    title=dict(text='Fertility Rate vs Life Expectancy',
               font=dict(family='Arial', size=24, color='black')),
    paper_bgcolor='white',
    plot_bgcolor='white',
    title_x=0.42,
    title_y=0.92,
    xaxis=dict(title='Fertility Rate', gridcolor='lightgray',
               title_font=dict(color='black')),
    yaxis=dict(title='Life Expectancy', gridcolor='lightgray',
               title_font=dict(color='black')),
    margin=dict(l=50, r=50, b=50, t=80),
)

fig.add_annotation(
    text='',
    x=0.5,
    y=0.5,
    showarrow=False,
    xref="paper",
    yref="paper",
    font=dict(color='black', size=16),
    xanchor='center',
    yanchor='auto',
    name='year_annotation'
)


fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 600
fig.layout.updatemenus[0].font.color = 'black'
fig.update_coloraxes(colorbar=dict(title_font=dict(color='black')))

def update_annotations(trace, points, selector):
    year = trace.hovertext[points.point_inds[0]]
    fig.update_annotations(text=f'Year: {year}',
    selector=dict(name='year_annotation'))

fig.for_each_trace(lambda trace: trace.on_hover(update_annotations))

fig.show()
```
