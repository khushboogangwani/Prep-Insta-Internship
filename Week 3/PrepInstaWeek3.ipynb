# **3. Outliers:**
Question: Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?




```python
z_score_threshold = 3

z_scores = np.abs((dataset1[['Age', 'Salary', 'Rating']] - dataset1[['Age', 'Salary', 'Rating']].mean()) / dataset1[['Age', 'Salary', 'Rating']].std())

data_no_outliers = dataset1[(z_scores <= z_score_threshold).all(axis=1)]

print("Dataset after removing outliers:")
print(data_no_outliers)
```

# **10. Handling Special Characters:**
Question: Examine all text-based columns (e.g., Location). Are there special characters or inconsistencies that need cleaning?




```python
text_columns = ['Location']


for column in text_columns:
    unique_values = dataset1[column].unique()
    print(f"\nUnique values in the '{column}' column:")
    print(unique_values)


dataset1['Location'] = dataset1['Location'].str.replace(r'[^a-zA-Z0-9\s,]', '', regex=True)
dataset1['Location'] = dataset1['Location'].str.strip()


print("\nDataset after cleaning text-based columns:")
print(dataset1)
```


# **11. Data Integrity:**
Question: Ensure data integrity by cross-referencing entries. For instance, does the Established column align with the Age column?




```python
# Assuming -1 is used for missing values in 'Age'
misaligned_entries = dataset1[dataset1['Age'] != -1]

# Filtering entries where 'Age' and 'Established' columns don't align
misaligned_entries = misaligned_entries[(misaligned_entries['Established'] != -1) & (misaligned_entries['Age'] - misaligned_entries['Established'] < 0)]

# Display entries where 'Age' and 'Established' columns don't align
print("Entries where 'Age' and 'Established' columns don't align:")
print(misaligned_entries[['Age', 'Established']])

# Replace missing 'Age' values with the difference between 'Established' and the minimum 'Established' value
dataset1.loc[dataset1['Age'] == -1, 'Age'] = dataset1['Established'] - dataset1['Established'].min()

# Display the dataset after handling misaligned entries
print("\nDataset after handling misaligned entries:")
print(dataset1)
```


# **12. Easy Apply Transformation:**
Question: If the Easy Apply column contains non-boolean values, how can it be transformed into a usable format?



```python
# Get unique values in the 'Easy Apply' column
unique_easy_apply_values = dataset1['Easy Apply'].unique()

print("Unique values in the 'Easy Apply' column:")
print(unique_easy_apply_values)

# Get the data type of the 'Easy Apply' column
easy_apply_data_type = dataset1['Easy Apply'].dtype

print("\nData type of the 'Easy Apply' column:", easy_apply_data_type)

# Check if the data type is not boolean and transform if needed
if easy_apply_data_type != bool:
    dataset1['Easy Apply'] = dataset1['Easy Apply'].map({'TRUE': True, 'FALSE': False})

# Display the dataset after transforming the 'Easy Apply' column
print("\nDataset after transforming 'Easy Apply' column:")
print(dataset1)
```

15. Consistent Rating Scale:
Question: Ensure a consistent rating scale in the Rating column. Should it be normalized or adjusted for uniform analysis?


```python
# Assuming 'Rating' is a numerical column
# Normalize the 'Rating' column to a common scale, e.g., between 0 and 1
min_rating = dataset1['Rating'].min()
max_rating = dataset1['Rating'].max()

dataset1['Normalized_Rating'] = (dataset1['Rating'] - min_rating) / (max_rating - min_rating)

# Display the dataset with the normalized 'Rating' column
print("\nDataset with normalized 'Rating' column:")
print(dataset1)
```
